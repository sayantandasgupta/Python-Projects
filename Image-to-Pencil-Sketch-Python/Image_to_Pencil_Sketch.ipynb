```python
from google.colab.patches import cv2_imshow
```

```python
#import the necessary module
import cv2
```

```python
#read the image from file
image = cv2.imread('friends.jpeg')
```

```python
cv2_imshow(image)
```

```python
# Converting BGR image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale_image)
```

```python
#invert the image to enhance features
inverted_image = 255 - grayscale_image
cv2_imshow(inverted_image)
```

```python
# blur the image to smoothen out sharper features
blurred_image = cv2.GaussianBlur(inverted_image, (21,21),0)
cv2_imshow(blurred_image)
```

```python
# invert the blurred image to get the original grayscale pixels
inverted_blurred = 255 - blurred_image
cv2_imshow(inverted_blurred)
```

```python
# finally to get the pencil sketch divide the grayscaled image with the inverted blurred image
pencil_sketch = cv2.divide(grayscale_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

```python
#save using cv2.imwrite() method
cv2.imwrite('sketch.jpg', pencil_sketch)
```
