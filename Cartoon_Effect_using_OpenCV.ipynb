# Cartoon Effect using OpenCV

There are very two simple steps through which any image can be converted into a cartoon image. First one is to apply a bilateral filter which is going to reduce the color palette, to render the cartoon effect. Second is to apply edge detection so that we can create sharp edged silhouttes.

Applying a bilateral filter is a lengthy process as compared to other filtering effects. A bilateral filter smooths flat regions while keeping the edges sharp.

Therefore to apply the bilateral filter effectively, we need to do the following steps:-

* Reduce the scale of the image
* Applying multiple short layers of bilateral filter

```
from google.colab.patches import cv2_imshow
```

```
import cv2

# Reading the Image 
image = cv2.imread("Sayantan.jpeg")
# Finding the Edges of Image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
gray = cv2.medianBlur(gray, 7) 
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 10)
# Making a Cartoon of the image
color = cv2.bilateralFilter(image, 12, 250, 250) 
cartoon = cv2.bitwise_and(color, color, mask=edges)
cv2.imwrite("res.jpg",cartoon)
#Visualize the cartoon image 
cv2_imshow(cartoon) 
```
